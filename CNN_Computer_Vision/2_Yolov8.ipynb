```python
# TREINAMENTO model
# passa como args arquitetura pronta ou customizada que quer usar
model = YOLO('yolov8n') # deteccao de obj base
```

```python
# FUN√á√ÉO DE TREINAMENTO .train
# No Treino, o modelo v√™ a imagem e a resposta (r√≥tulo). Ele decora padr√µes.
# cria .yaml
model.train(data='custom_dataset.yaml',
           # Argumentos
            epochs=30,
            patience=8,
            batch=16,  # se -1 valor otimiza para GPU
            imgsz=640, # necarr√°rio ser / por 32
            workers=8,
            pretrained=True,
            resume=False,  # retomar o treinamento do √∫ltimo ponto de verifica√ß√£o
            single_cls=False,  # Indica se todas as classes ser√£o iguais
            # project='runs/detect',  # Default = /home/{user}/Documents/ultralytics/runs
            box=7.5,  # Maior recall, melhor IoU, menor precission,
            cls=0.5,  # Bbox class melhorada
            dfl=1.5,  # melhores limites do bbox
            val=True, # se executa valida√ß√£o apos cada √©poca (-desempenho)
            degrees=0.3,
            hsv_s=0.3,
            hsv_v=0.3,
            scale=0.5,
            fliplr=0.5
            )
```

```python
# VALIDA√á√ÉO DE TESTES
# local onde esta o modelo treinado
model = YOLO('runs/detect/train/weights/best.pt')
```

```python
# FUN√á√ÉO DE VALIDA√á√ÉO  .val
results = model.val(project='runs/detect',
                    imgsz=640,
                    batch=16,
                    conf=0.001,
                    iou=0.7,
                    save_json=False,  # Salvar em JSON {image_id, cls, bbox, conf} de cada imagem no conjunto de dados
                    save_hybrid=False,  # Bounding box labels + inference on the output image (Ultralytics 8.0.178 = results are wrong when this argmunent is True)
                    split='test')  # train, val or test
# results
# results.box
# results.box.map50
```
