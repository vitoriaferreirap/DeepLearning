## Fundamentos Matematicos


```python
import torch
```

```python
# criando um vetor com 3 elementos
vetor = torch.tensor([1.,2.,3.])
print("Vetor: ", vetor)
```

```python
# criando uma matriz de 2x3
matriz = torch.tensor([[1.,2.,3.,],
                       [4.,5.,6.,]])
print("Matriz:\n", matriz)
```

```python
# Produto escalar - multiplicacao de dois vetores seguido de soma
# palavra chave dot

#vetore
v1 = torch.tensor([1.,2.,3.])
v2 = torch.tensor([10.,9.,8])

#produto escalar
produto_escalar = torch.dot(v1,v2)
print("Produto escalar: ", produto_escalar)
```

```python
# Multiplicacao Matriz x Vetor - palavra chave matmul

#matriz 2x3
m = torch.tensor([[1., 2., 3.], #soma 4
                  [4., 5., 6.]]) #soma 10

#vetor 3d
v = torch.tensor([1., 0., 1.])

#multiplicacao
res = torch.matmul(m, v)
print("matriz x vetor: ", res)

```

```python
# Tensor 3D, simulando uma imagem RGB 244x244
# - 3 canais: R (vermelho), G (verde), B (azul)
# - 244x244 pixels por canal
# - torch.zeros cria um tensor preenchido com zeros
#   (imagem "preta" inicialmente)

img = torch.zeros((3, 244, 244))
print("Tensor de imagem:", img.shape)  # Mostra as dimensões (3, 244, 244)

# Acessando o canal vermelho da imagem
# - img[0] seleciona o primeiro canal (R)
# - resultado é um tensor 2D (altura x largura)
canal_vermelho = img[0]
print("Canal vermelho shape:", canal_vermelho.shape)
```
